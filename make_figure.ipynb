```python
df = pd.read_csv("./Dataset/X_240905.csv", index_col=0, encoding='utf-8', low_memory=False)
```

```python
df.head(10)
```

```python
y = pd.read_csv("./Dataset/y_240905.csv", index_col=0, encoding='utf-8', low_memory=False)
y.head(10)
```

```python
# 선택할 컬럼 리스트
selected_columns = [
    'AS1_JOBB', 
    'AS1_DRINK', 
    'AS1_EDUA', 
    'AS1_SMOKEA', 
    'AS1_PHYACTL', 
    'AS1_PHYACTM', 
    'AS1_PHYACTH', 
    'AS1_HEALTH'
]

# 선택한 컬럼들로 새로운 데이터프레임 생성
selected_df = df[selected_columns]

# 'CVD' 값에 따라 그룹 나누기
group_0 = selected_df[y['CVD'] == 0]
group_1 = selected_df[y['CVD'] == 1]

# 각 컬럼의 값 별 개수를 담을 데이터프레임 생성
value_counts_list = []

for column in selected_df.columns:
    total_counts = selected_df[column].value_counts().reset_index()
    group_0_counts = group_0[column].value_counts().reindex(total_counts['index'], fill_value=0).reset_index()
    group_1_counts = group_1[column].value_counts().reindex(total_counts['index'], fill_value=0).reset_index()

    total_counts.columns = ['Value', 'Total_Count']
    group_0_counts.columns = ['Value', 'Group_0_Count']
    group_1_counts.columns = ['Value', 'Group_1_Count']

    counts = pd.DataFrame({
        'Column_Value': column + '_' + total_counts['Value'].astype(str),
        'Total_Count': total_counts['Total_Count'],
        'Group_0_Count': group_0_counts['Group_0_Count'],
        'Group_1_Count': group_1_counts['Group_1_Count']
    })

    value_counts_list.append(counts)

# 모든 결과를 하나의 데이터프레임으로 통합
value_counts_df = pd.concat(value_counts_list, ignore_index=True)

# 'Column_Value' 기준으로 정렬
sorted_value_counts_df = value_counts_df.sort_values(by='Column_Value').reset_index(drop=True)

# 결과 출력
print(sorted_value_counts_df)
```

```python
column_means = df.mean().round(2)

# 조건에 따라 데이터를 두 그룹으로 나누기
group_0 = df[y['CVD'] == 0]
group_1 = df[y['CVD'] == 1]

# 각 그룹의 평균 계산
group_0_means = group_0.mean().round(2)
group_1_means = group_1.mean().round(2)

results = pd.DataFrame(
    {
        'Overall (n = 9622) m': column_means,
        'Overall (n = 9622) s': df.std().round(2),
        'CVD (n = 9335) m': group_0_means,
        'CVD (n = 9335) s': group_0.std().round(2),
        'non CVD (n = 287) m': group_1_means,
        'non CVD (n = 287) s': group_1.std().round(2)
    },
    index=df.columns
)
results.head()
```

```python
results.to_csv("./Dataset/result/mean.csv")
```

```python
# 각 컬럼에 대해 0과 1의 개수 계산
zero_counts = (df == 1).sum()
one_counts = (df == 2).sum()

# 결과를 데이터프레임으로 통합
results_bin = pd.DataFrame({
    'Overall (n = 9622) 0': zero_counts,
    'Overall (n = 9622) 1': one_counts,
    'CVD (n = 9335) 0': (group_0 == 1).sum(),
    'CVD (n = 9335) 1': (group_0 == 2).sum(),
    'non CVD (n = 287) 0': (group_1 == 1).sum(),
    'non CVD (n = 287) 1': (group_1 == 2).sum()
})
results_bin.head()
```

```python
results_bin.to_csv("./Dataset/bin.csv")
```
