# 베이스라인 만들기 대작전

- var_to_use 이용하여 연속형, 범주형 구분하여 작성할 것

```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
data = pd.read_csv('./Dataset/result_240905.csv', index_col=0)

# 결과를 저장할 데이터프레임 생성
results = []

data.head()
```

```python
# 각 독립 변수에 대해 통계치를 계산하고 p-value를 구합니다.
for column in data.columns:
    if column == 'CVD':  # 종속 변수는 생략
        continue

    # 'CVD' 변수가 1인 경우와 0인 경우로 나눕니다.
    cvd_true = data[data['CVD'] == 1][column]
    cvd_false = data[data['CVD'] == 0][column]

    # 통계치 계산
    mean = data[column].mean()
    var = data[column].var()
    min_value = data[column].min()
    max_value = data[column].max()
    
    # t-검정 수행
    _, p_value = ttest_ind(cvd_true, cvd_false, equal_var=False)

    # 결과 저장
    results.append({
        '칼럼의 이름': column,
        '칼럼의 설명': '',
        '평균': mean,
        '분산': var,
        '최소값': min_value,
        '최대값': max_value,
        'p-value': p_value
    })

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)
```
